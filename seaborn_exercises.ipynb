```python
def get_db_url(data_base):
    return (f'mysql+pymysql://{env.username}:{env.password}@{env.host}/{data_base}')
    
```

```python
orders = pd.read_sql("select * from orders", get_db_url("chipotle"))
```

```python
orders
```

```python
popular_items = orders[["item_name" ,"quantity" , "item_price"]]

popular_items.item_price = popular_items.item_price.str.strip('$').astype(float)

popular_items
```

```python
popular_items.groupby("item_name", as_index=False)[["item_price"]].sum().sort_values("item_price", ascending=False).head(4)
```

```python
sns.catplot(data=popular_items.groupby("item_name", as_index=False)[["item_price"]].sum().sort_values("item_price", ascending=False).head(4), x="item_name",  y="item_price" ,kind="bar")
plt.show()
```
